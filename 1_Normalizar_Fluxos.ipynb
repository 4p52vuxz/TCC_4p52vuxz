```python
import pandas as pd
import numpy as np
```

```python
path = 'Combinacao_Resultados.xlsx'
df = pd.read_excel(path, header=3, sheet_name='Fluxos')
df
```

```python
cols_1 = ['Jacobi_Fluxo_Grupo1', 'Gauss_Fluxo_Grupo1']
cols_2 = ['Jacobi_Fluxo_Grupo2', 'Gauss_Fluxo_Grupo2']
max_1 = df[cols_1].max().to_numpy()
df[cols_1] = df[cols_1].div(max_1)
df[cols_2] = df[cols_2].div(max_1)
df['Posicao'] = df['Posicao'] - 50
df.to_csv('Combinacao_Resultados_Normalizado.csv', index=False)
df
```
